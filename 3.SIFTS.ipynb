### Get all xml.gz from SIFTS

```python
# pip install pandas 
import re, os, zlib, time, json, pickle, sqlite3, urllib, urllib.request, subprocess, datetime, multiprocessing, pandas as pd
con = sqlite3.connect('/scratch/xg590/scratch/sifts.20220302.db')  
cur = con.cursor()
```

```python
pd.read_sql("SELECT name FROM sqlite_master WHERE type ='table';", con)
```

```python
from IPython.core.display import display, HTML, SVG
import pandas as pd
display(HTML("<style>.container { width:100% !important; }</style>"))  
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None) 
df = pd.read_pickle('data/df_latest.pkl') 
df.head()
```

```python
df['pdb_id_lower']     = df.apply(lambda x: x.pdb_id.lower(), axis=1)  
df['res_num_extented'] = df.apply(lambda r: r.covalent_bond_record.split(',')[6] if r.covalent_bond_record.split(',')[9]=='?' else r.covalent_bond_record.split(',')[6] + r.covalent_bond_record.split(',')[9], axis=1)  
df[['pdb_id_lower','chain_id','res_num_extented']].to_sql('pdb_cov',con,if_exists="replace",index=True)
con.commit()
```

```python
cur.execute("SELECT * FROM pdb_cov LIMIT 3;")
cur.fetchall()
```

```python
df_hit = pd.read_sql_query('''SELECT sifts.*       ,         pdb_cov.adduct_id 
                              FROM   sifts       JOIN        pdb_cov 
                              ON     sifts.pdb_accessionid = pdb_cov.pdb_id_lower 
                                 AND sifts.pdb_chainid     = pdb_cov.chain_id 
                                 AND sifts.pdb_resnum      = pdb_cov.res_num_extented''', con)
```

```python
len(df_hit)
```

```python
df_hit.to_csv(datetime.datetime.fromtimestamp(time.time()).strftime("data/sifts.matched.csv"),index=False)
```
