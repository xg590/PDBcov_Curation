```python
import urllib, zlib
from bs4 import BeautifulSoup
def sifts_get_mapping_by_pdbid(p): 
    with open(p, 'rb') as fr: 
        xml  = zlib.decompressobj(32+zlib.MAX_WBITS).decompress(fr.read()) 
        soup = BeautifulSoup(xml, "lxml")
        s    = ''
        for residue in soup.find_all('residue'): 
            UniProt = residue.find(dbsource="UniProt")
            PDB = residue.find(dbsource="PDB") 
            try:
                if UniProt:
                    s += '{},{},{},{},{},{},{}\n'.format(PDB['dbaccessionid'], PDB['dbchainid'], PDB['dbresnum'], PDB['dbresname'], UniProt['dbaccessionid'], UniProt['dbresnum'], UniProt['dbresname'])
                else:
                    s += '{},{},{},{},null,null,null\n'.format(PDB['dbaccessionid'], PDB['dbchainid'], PDB['dbresnum'], PDB['dbresname']) 
            except Exception as e:
                print(pdbid, residue)
                s += json.dumps(residue.attrs)
                s += '\n'
        return s

sifts_get_mapping_by_pdbid('data/sifts/102l.xml.gz')[:60]
```

```python
from IPython.core.display import display, HTML, SVG
import pandas as pd
display(HTML("<style>.container { width:100% !important; }</style>"))  
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None) 
df = pd.read_pickle('data/df_latest.pkl') 
df.head()
```

```python
df['pdb_id_lower']     = df.apply(lambda x: x.pdb_id.lower(), axis=1)  
df['res_num_extented'] = df.apply(lambda r: r.covalent_bond_record.split(',')[6] if r.covalent_bond_record.split(',')[9]=='?' else r.covalent_bond_record.split(',')[6] + r.covalent_bond_record.split(',')[9], axis=1)  
df[['pdb_id_lower','chain_id','res_num_extented']].to_sql('pdb_cov',con,if_exists="replace",index=True)
con.commit()
```

```python
cur.execute("SELECT * FROM pdb_cov LIMIT 3;")
cur.fetchall()
```

```python
df_hit = pd.read_sql_query('''SELECT sifts.*       ,         pdb_cov.adduct_id 
                              FROM   sifts       JOIN        pdb_cov 
                              ON     sifts.pdb_accessionid = pdb_cov.pdb_id_lower 
                                 AND sifts.pdb_chainid     = pdb_cov.chain_id 
                                 AND sifts.pdb_resnum      = pdb_cov.res_num_extented''', con)
```
